```python
# Get list of genres
genres = df_main['artist_genres'].unique().tolist()
genres = [i.strip("[]").split(', ') for i in genres]
genres = [item for sublist in genres for item in sublist]
genres = [x.strip("''") for x in genres]
genres = list( dict.fromkeys(genres) )
genres = list(filter(None, genres))
```

```python
dff = df_main[['track_id','artist_genres','track_popularity']]
dff
```

```python
from ast import literal_eval
from itertools import chain
from collections import Counter
```

```python
# flatten lists in genres column?
dff['artist_genres'] = dff['artist_genres'].map(literal_eval)
```

```python
# separate genres (rows are copied, meaning there're multiple rows for same tracks now)
lst_col = 'artist_genres'

dff = pd.DataFrame({col:np.repeat(dff[col].values, dff[lst_col].str.len())for col in dff.columns.drop(lst_col)}).assign(**{lst_col:np.concatenate(dff[lst_col].values)})[dff.columns]
```

```python
vars_ind_onehot = []

df_all_onehot = df_all.copy()

# Choose a column name
col = 'artist_genre'

for col in vars_ind_toonehot:
    print(col)
    
    # use pd.get_dummies on  df_all[col] 
    df_oh = pd.get_dummies(df_all[col], drop_first=False)
    
    # Find the column name of the most frequent category
    col_mostFreq =  df_oh.sum(axis = 0).idxmax()
    
    # Drop the column of the most frequent category
    df_oh = df_oh.drop(col_mostFreq, axis=1)
        
    # Rename the columns to have the original variable name as a prefix
    oh_names = df_oh.columns = col + '_' + df_oh.columns
    df_oh.columns = oh_names
    
    df_all_onehot = pd.concat([df_all_onehot, df_oh], axis = 1, sort = False)

    del df_all_onehot[col]
    vars_ind_onehot.extend(oh_names)
```
